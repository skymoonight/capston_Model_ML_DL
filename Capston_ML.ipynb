```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,f1_score,precision_score
from sklearn import svm
import pandas as pd

data1 = pd.read_csv('./CIC_RF_train.csv')
data2 = pd.read_csv('./CIC_RF_test.csv')

X_train = data1.iloc[:,:-1]
y_train = data1.iloc[:,-1]
X_test = data2.iloc[:,:-1]
y_test = data2.iloc[:,-1]

X_train
```

```python
X_test
```

```python
neighbors_value=2
KNN_classifier = KNeighborsClassifier(n_neighbors=neighbors_value)

KNN_classifier.fit(X_train,y_train)

result = KNN_classifier.predict(X_test)

res = classification_report(y_test,result)
print(res)
```

```python
tree = DecisionTreeClassifier(random_state=0,min_samples_leaf=3)

tree.fit(X_train,y_train)

result2 = tree.predict(X_test)
res2 = classification_report(y_test,result2)
print(res2)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=0, max_features=10, criterion='gini',min_samples_leaf=3)

rf.fit(X_train,y_train)

result3 = rf.predict(X_test)
res3 = classification_report(y_test,result3)
print(res3)
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(X_train,y_train)

result4 = clf.predict(X_test)
res4 = classification_report(y_test,result4)
print(res4)
```
